```python
import requests
import bs4
import json
import re
```

```python
request=requests.get("https://www.moneycontrol.com/company-article/relianceindustries/news/RI")
```

to see content of our request varuable

```python
print(request.text)
```

conversion of raw text to html

```python
soup=bs4.BeautifulSoup(request.text,"html.parser")
```

```python
print(type(soup))
```

getting title of the webpage

```python
print(soup.title)
```

selecting a paragraph

```python
para = soup.select('p')
```

```python
print(para)
```

```python
anchor = soup.find_all('a', attrs={'class': 'arial11_summ'})
print(anchor)
```

extracting all the title and urls with the help  of anchor tag

```python
for attr in anchor:
  print('Title :', attr['title'])
  print('URL :', attr['href'])
  print('--------------------------------')
```

extracting information from news blog

1.headline
2.short description
3.published date and time
4.content of block
5.URLS
6.tags



```python
all_script = soup.find_all('script', attrs={'type': 'application/ld+json'})
type(all_script)
```

```python
len(all_script)
```

```python
print(all_script[0])
```

removing all spaces and special charcters(converting string into dictionary)

```python
raw_article_str=all_script[0].get_text().replace('\r\n',' ')
print(raw_article_str)
```

```python
parts=re.split('[\n\t]',raw_article_str)
print(parts)
```

```python
parts[::2]=map(lambda s: "".join(s.split()),parts[::2])
print(parts)
```

```python
article_str = parts[1:]
article_str = article_str[:-1]
print(article_str)
```

```python
article_str_string=" ".join(article_str)
if article_str_string:
  article_dict=json.loads(article_str_string)
  print(article_dict)
else:
  print("No article found")
```

extracting tags

```python
all_tags=soup.find_all('div',attrs={'class':'tags_first_line'})
print(all_tags)
```

```python
lst_all_tags=[]
for attr in all_tags:
  lst_all_tags.append(attr.text)
print(lst_all_tags)
```

```python
tags=lst_all_tags[0].replace('Tags : ','')
tags=tags.replace('\n','')
print(tags)
```
